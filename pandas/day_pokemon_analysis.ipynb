```python
import pandas as pd

df =pd.read_csv("pokemon.csv", index_col="name")

print (df)
```

```python
print(df.head())
print(df.tail())
```

```python
print(f"Rows: {df.shape[0]}")
print(f"Column: {df.shape[1]}")
print(df.columns.tolist())
print(df.dtypes)
print(df.info())
```

```python
print(df.isnull().to_string())
count_missing=df[df["type_2"].isnull()]
print(f"Pokemon with no type2 {count_missing}")
```

```python
fire_type=df[df["type_1"]=="Fire"]
print(fire_type)
attcking_pokemon=df[df["attack"]>100]
print(attcking_pokemon)
df['total']=df['attack']+df['defense']+df['speed']+df["sp_atk"]+df['sp_def']
print(df.head())
high_stat=(df[df['total']>300])
print(high_stat)
```

```python

print((df["hp"].mean()))
print((df["attack"].mean()))
print((df["defense"].mean()))

highest_attack_row = df[df['attack'] == df['attack'].max()]

print(highest_attack_row)

slowest_pokemon=df[df['speed']==df['speed'].min()]
print(slowest_pokemon)

type_counter=df['type_1'].value_counts()
print(type_counter)
```

```python
group=df.groupby("type_1")
print(group["attack"].mean())
print(group["defense"].mean())
print(group["speed"].mean())

avg_speed = df.groupby('type_1')['speed'].mean()
highest_type = avg_speed.idxmax()
highest_value = avg_speed.max()
print(f"The Type 1 with the highest average Speed is {highest_type} with {highest_value:.2f}")
print(avg_speed.sort_values(ascending=False))
```
